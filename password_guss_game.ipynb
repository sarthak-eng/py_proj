```python
import random
```

```python
#giving words to make password
easy_words=['apple','ball','cat','dog']
medium_words=['python','java','javascript','pandas']
hard_words=['train','executive','hack','content']
```

```python
print('Welcome to password generator')
print("choose difficulty level")
```

```python
level=input('enter difficulty level:').lower()
if level=='easy':
  secret=random.choice(easy_words)
elif level=='medium':
  secret=random.choice(medium_words)
elif level=='hard':
  secret=random.choice(hard_words)
else:
  print('invalid input')
```

```python
#3.game loop
attempts=0
print('/nguess the password')

while True:
  guess=input("enter your guess:").lower()
  attempts+=1

  if guess==secret:
    print(f'congrats!,your guess in {attempts} is write')
    break
```

```python
#4.hint to show which letter is correct
if guess<secret:
  print('your guess is too low')
else:
   print('your guess is too high')
```
