```python
import random
```

```python
subjects=[
    "sharukh khan",
    "virath kholi",
    "nirmila sitaraman",
    "A mumbai cat",
    "Group of monkeys",
    "PM Modi",
    "AUTO driver",
    "anaconda"
]

actions=[
    "launches",
    "cancels",
    "dance with",
    "eats",
    "declear war on",
    "orders",
    "celebeates",
    "found"

]

places_or_things=[
    "at red fort",
    "in mumbai local train",
    "a plate of samosa",
    "inside parliament",
    "at ganga ghat",
    "WIN THE  IPL",
    "inside the white house"

]
```

```python
from re import sub
#3.start the headline generation loop
while True:
  subject=random.choice(subjects)
  action=random.choice(actions)
  place_or_thing=random.choice(places_or_things)

  headline=f"BREAKING NEWS:{subject} {action} {place_or_thing}"
  headline=headline.title()
  print(headline)
  repeat=input("Generate another headline or NOT?").strip().upper()
  if repeat=="NOT":
    break

```
