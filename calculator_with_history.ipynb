```python
import re

HISTORY_FILE="hist.txt"

def show_hist():
  file=open(HISTORY_FILE,"r")
  lines=file.readlines()
  if len(lines)==0:
    print("No history found")
  else:
      for line in reversed(lines):
        print(line.strip())

  file.close()

def save_hist(eq,result):
  file=open(HISTORY_FILE,"a")
  file.write(eq+"="+str(result)+"\n")
  file.close()

def calc(user_input):
  # Use regex to find numbers and operator
  match = re.match(r'(\d+\.?\d*)\s*([-+*/])\s*(\d+\.?\d*)', user_input)
  if not match:
      print("Invalid input")
      return

  a = float(match.group(1))
  op = match.group(2)
  b = float(match.group(3))

  if op=="+":
    result=a+b
  elif op=="-":
    result=a-b
  elif op=="*":
    result=a*b
  elif op=="/":
    if b==0:
      print("Division by zero is not allowed")
      return
    result=a/b
  else:
    print("Invalid operator")
    return
  if result.is_integer():
    result=int(result)
  print(f"{a} {op} {b} = {result}")
  save_hist(user_input,result)

def main():
  print("simple calculator(type history,clear,exir)")
  while True:
    user_input=input("calc{+,-,/,*},command{hist,clear,exit}")
    if user_input.lower()=="history":
      show_hist()
    elif user_input.lower()=="clear":
      file=open(HISTORY_FILE,"w")
      file.close()
      print("history cleared")

    elif user_input.lower()=="exit":
      print("exiting")
      break
    else:
      calc(user_input)

main()
```
